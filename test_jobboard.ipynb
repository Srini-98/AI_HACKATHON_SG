## Supabase Table

```python
from dotenv import load_dotenv
import os
import supabase
from supabase import create_client, Client


load_dotenv()

# Access the environment variables from the .env file
url = os.environ.get('SUPABASE_URL')
key = os.environ.get('SUPABASE_KEY')
client = create_client(url, key)
```

```python
response = client.table('jobs').select('*').execute()
df = pd.DataFrame(response.data)
```

```python
df
```

```python
from langchain.llms import Ollama

def generate_job_description(role, company):
    ollama = Ollama(base_url="http://localhost:11434", model="llama3")

    prompt = f"Generate a job description for the role of {role} at {company}."
    response = ollama(prompt=prompt)
    return response

# Generate job descriptions
df['description'] = df.apply(lambda row: generate_job_description(row['Role'], row['Company']), axis=1)
```
